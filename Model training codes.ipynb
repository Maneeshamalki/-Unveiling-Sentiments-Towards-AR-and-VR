```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import pandas as pd

#dataset
Data =pd.read_excel('C:/Users/Admin/Desktop/Internship SLT/Research - AR ,VR/Responses.xlsx')


# Sample dataset (you should replace this with your own dataset)
comments = Data['Educational_Benefits']
labels = Data['Rating_educational ']  

comments = [comment for comment in comments if isinstance(comment, str)]
labels = labels[:len(comments)]

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(comments, labels, test_size=0.2, random_state=42)

# Initialize the CountVectorizer
vectorizer = CountVectorizer()

# Transform the text data into numerical features
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Initialize and train the Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train_vec, y_train)

# Make predictions on the test set
predictions = classifier.predict(X_test_vec)

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

```

```python
# New comments to predict
new_comments = Data['Educational_Benefits']

new_comments = [comment for comment in new_comments if isinstance(comment, str)]

# Transform the new comments into numerical features using the same vectorizer
new_comments_vec = vectorizer.transform(new_comments)




# Make predictions on the new comments
new_predictions = classifier.predict(new_comments_vec)

# Convert the numerical predictions back to sentiment labels
sentiment_labels = ["Negative", "Positive", "Neutral"]
predicted_sentiments = [sentiment_labels[prediction] for prediction in new_predictions]

# Print the predicted sentiments for the new comments
for comment, sentiment in zip(new_comments, predicted_sentiments):
    print(f"Comment: {comment}\nPredicted Sentiment: {sentiment}\n")
```
